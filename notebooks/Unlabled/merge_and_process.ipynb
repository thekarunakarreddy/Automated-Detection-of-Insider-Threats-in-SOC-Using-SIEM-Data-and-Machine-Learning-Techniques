```python
import pandas as pd

# Load daily feature files
logon_df = pd.read_csv('../outputs/daily_user_logons_enriched.csv')
file_df = pd.read_csv('../outputs/daily_user_files_enriched.csv')
device_df = pd.read_csv('../outputs/daily_user_devices_enriched.csv')
email_df = pd.read_csv('../outputs/daily_user_email_enriched.csv')
http_df = pd.read_csv('../outputs/daily_user_http_enriched.csv')

# Merge all on ['user','date_only']
merged_df = logon_df.merge(file_df, on=['user', 'date_only'], how='outer')
merged_df = merged_df.merge(device_df, on=['user', 'date_only'], how='outer')
merged_df = merged_df.merge(email_df, on=['user', 'date_only'], how='outer')
merged_df = merged_df.merge(http_df, on=['user', 'date_only'], how='outer')

# Fill missing numeric columns with 0
merged_df = merged_df.fillna(0)

# Load psychometric data
psych_df = pd.read_csv(
    r"C:\Users\karun\OneDrive\Documents\RIK\data\CERT-dataset\psychometric.csv"
)

# Merge psychometric traits per user
merged_df = merged_df.merge(
    psych_df,
    left_on='user',
    right_on='user_id',
    how='left'
)

# Drop duplicate columns if present
merged_df.drop(
    columns=['user_id_y', 'employee_name_x', 'employee_name_y'],
    errors='ignore',
    inplace=True
)

# Save final merged data
merged_df.to_csv(
    '../outputs/RIK_final_training_data.csv',
    index=False
)

merged_df.head()
```
