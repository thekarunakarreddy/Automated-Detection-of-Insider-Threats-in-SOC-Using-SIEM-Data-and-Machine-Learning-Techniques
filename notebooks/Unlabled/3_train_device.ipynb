```python
import pandas as pd

device_df = pd.read_csv(
    r"C:\Users\karun\OneDrive\Documents\RIK\data\CERT-dataset\device.csv"
)

device_df['date'] = pd.to_datetime(device_df['date'], errors='coerce')
device_df['date_only'] = device_df['date'].dt.date

device_df['is_connect'] = device_df['activity'].str.lower().str.contains('connect').astype(int)
device_df['is_disconnect'] = device_df['activity'].str.lower().str.contains('disconnect').astype(int)

# Fill NaNs in file_tree
device_df['file_tree'] = device_df['file_tree'].fillna('')

device_df['is_external_path'] = device_df['file_tree'].str.contains(r'^[A-Z]:\\', regex=True).astype(int)

daily_user_devices = device_df.groupby(
    ['user', 'date_only'], as_index=False
).agg(
    device_events_per_day=('id', 'count'),
    unique_pcs_per_day=('pc', pd.Series.nunique),
    unique_filetrees_per_day=('file_tree', pd.Series.nunique),
    connect_events_per_day=('is_connect', 'sum'),
    disconnect_events_per_day=('is_disconnect', 'sum'),
    external_path_access_per_day=('is_external_path', 'sum')
)

daily_user_devices.to_csv(
    '../outputs/daily_user_devices_enriched.csv',
    index=False
)

daily_user_devices.head()
```
